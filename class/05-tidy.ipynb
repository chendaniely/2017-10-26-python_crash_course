```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather.head()
```

```python
weather_melt = pd.melt(weather,
                      id_vars=['id', 'year', 'month', 'element'],
                      var_name='day',
                      value_name='temp')
```

```python
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
```

```python
weather_tidy = weather_tidy.reset_index()
```
